```python
import requests
import PyPDF2
import os
import pandas as pd

def read_pdf_from_url(pdf_url):
    try:
        # Suppressing requests warnings
        requests.packages.urllib3.disable_warnings(requests.packages.urllib3.exceptions.InsecureRequestWarning)

        response = requests.get(pdf_url, verify=False)
        with open("temp.pdf", "wb") as pdf_file:
            pdf_file.write(response.content)

        pdf_text = ""
        with open("temp.pdf", "rb") as pdf_file:
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            total_pages = len(pdf_reader.pages)

            for page_num in range(total_pages):
                page = pdf_reader.pages[page_num]
                pdf_text += page.extract_text()

        return pdf_text
    except Exception as e:
        print(f"Error: {e}")
        return None
    finally:
        # Cleanup: Remove temporary PDF file
        try:
            os.remove("temp.pdf")
        except OSError:
            pass

# Manually adding URL list:
manual_urls = [
    "https://www.indiabudget.gov.in/doc/bspeech/bs194748.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs194849.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs194950.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs195051.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs195152.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs195253.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs195354.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs195455.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs195556.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs195657.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs195758.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs195859.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs195960.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs196061.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs196162.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs196263.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs196364.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs196465.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs196566.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs196667.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs196768.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs196869.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs196970.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs197071.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs197172.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs197273.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs197374.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs197475.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs197576.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs197677.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs197778.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs197879.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs197980.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs198081.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs198182.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs198283.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs198384.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs198485.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs198586.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs198687.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs198788.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs198889.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs198990.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs199091.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs199192.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs199293.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs199394.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs199495.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs199596.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs199697.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs199798.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs199899.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs19992000.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs200001.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs200102.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs200203.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs200304.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs200405.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs200506.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs200607.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs200708.pdf",
    "https://megfinance.gov.in/files/budget-speech/budget_speech-2008-2009.pdf",
    "https://megfinance.gov.in/files/budget-speech/budget_speech-2009-2010.pdf",
    "https://megfinance.gov.in/files/budget-speech/budget_speech-2010-2011.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs201112.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs201213.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs201314.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs201415.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs201516.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs201617.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs201718.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs201819.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs201920.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs202021.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs202122.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs202223.pdf",
    "https://www.indiabudget.gov.in/doc/bspeech/bs2023_24.pdf",
]

dfs = [] # Empty list to store DataFrames

for pdf_url in manual_urls:
    pdf_text = read_pdf_from_url(pdf_url)

    if pdf_text:
        # Creating a DataFrame for each PDF
        df = pd.DataFrame({"Text": [pdf_text], "Filename": [os.path.basename(pdf_url)]})
        dfs.append(df)
    else:
        print(f"Failed to read the PDF from {pdf_url}.")

# Concatenating all DataFrames into a single DataFrame
df = pd.concat(dfs, ignore_index=True)

print(df) # Printing the DataFrame

# Saving the DataFrame to a JSON file for later use
df.to_json('output_dataset.json', orient='records', lines=True)
```
